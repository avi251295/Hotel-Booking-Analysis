### Data Analysis and visuliazation

```python
df
```

```python
len(df)
```

```python
# find cancellation percentage
cancel_per = df['is_canceled'].value_counts(normalize=True)
cancel_per
```

```python
plt.title('Reservation status count')
plt.bar(['Not_Canceled','Canceled'],df['is_canceled'].value_counts(), edgecolor='k')
plt.show()
```

```python
sns.countplot(x='hotel', hue='is_canceled', data=df)
plt.title('Reservation Status in different hotel')
plt.ylabel('number of reservation')
plt.xlabel('Hotel')
plt.show()
```

```python
# percenrage value of cancelation in resort hotel.
resort_hotel = df[df['hotel']=='Resort Hotel']['is_canceled'].value_counts(normalize=True)
resort_hotel
```

```python
# percenrage value of cancelation in city hotel.
City_hotel = df[df['hotel']=='City Hotel']['is_canceled'].value_counts(normalize=True)
City_hotel
```

```python
# Variation of mean price
re_hotel=df[df['hotel']=='Resort Hotel'].groupby('reservation_status_date')[['adr']].mean()
ci_hotel=df[df['hotel']=='City Hotel'].groupby('reservation_status_date')[['adr']].mean()
```

```python
#Visualization of 'Average daily rate in city and resort hotel'
plt.figure(figsize=(15,8))
plt.plot(re_hotel.index, re_hotel['adr'], label='Resort Hotel')
plt.plot(ci_hotel.index, ci_hotel['adr'], label='Resort Hotel')
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=20)
plt.legend(fontsize=10)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
df
```

```python
#Reservation status per month
plt.figure(figsize=(12, 6))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df)

# Set the title and axis labels
plt.title('Reservation Status per Month', size=15)
plt.xlabel('Month', size=12)
plt.ylabel('Number of Reservations', size=12)

# Update the legend
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))

plt.show()

#here we observe that highest cancelation in Jan whereas lowest cancelation in month of Aug. But why let's check with price
```

```python
ADR_per_month= df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
ADR_per_month
```

```python
# let's visulize variation of price per month'
plt.figure(figsize=(12, 6))
plt.title('ADR per month')
sns.barplot(x='month', y='adr', data=ADR_per_month)
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
# Yes, it is correct that price in month of jan is highest whereas in month of aug is lowest 
#so price might be a reason of more number cancelation in month of jan
```

```python
cancelled_data = df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
top_10_country
```

```python
plt.figure(figsize=(15,7))
plt.title('Top 10 country with reservation cancel',size=15)
plt.pie(top_10_country, autopct='%.2f', labels= top_10_country.index)
plt.show()

# Here got that portugal country have highest percentage of cancellation. This can be handle by controlling price, advertizement, permotional event.
```

```python
# visualize effect of price on cancellation and not cancellation

cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data= df[df['is_canceled']==0]
not_cancelled_data_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace=True)
not_cancelled_data_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(15,8))
plt.title('Average Daily Rate', size=20)
plt.plot(not_cancelled_data_adr['reservation_status_date'],not_cancelled_data_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
plt.show()
```
